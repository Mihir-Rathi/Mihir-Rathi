### t1.py

```python
import numpy as np

n=np.array([
    [72,85,60,78],
    [88,92,95,90],
    [45,52,48,55],
    [65,70,72,68]
])

total_student=n.sum(axis=1)
avg_student=n.mean(axis=1)

print(f"Total Marks per Student:{total_student}")
print(f"Average Marks per Student:{avg_student}")

avg_subject=n.mean(axis=0)
print(f"Average Marks per Subject:{avg_subject}")

classification=np.where(avg_student>=80,"Distinction",
                        np.where(avg_student>=65,"Merit",
                                 np.where(avg_student>=50,"Pass","Fail")))

print(f"\nClassification:\n{classification}")
```

### t2.py

```python
import numpy as np

sales = np.random.randint(100, 1000, (12, 4))

print("Sales Data:\n", sales)

month_total = np.sum(sales, axis=1)
best_month = np.argmax(month_total) + 1

print("\nBest Selling Month:", best_month)

product_total = np.sum(sales, axis=0)
worst_product = np.argmin(product_total) + 1

print("Worst Product:", worst_product)

good_months = np.where(np.all(sales > 500, axis=1))[0] + 1

print("Months where all products sold more than 500 units:")
print(good_months)
```

### t4.py

```python
import numpy as np

def my_range(start,stop,step):
    
    n=int((stop-start)/step)
    a=np.zeros(n)
    
    for i in range(n):
        a[i]=start+i*step
        
    return a

x=my_range(2,10,2)
print(x)
```

### t5.py

```python
import numpy as np

a = np.random.randint(1,100,(10,10))

print("Matrix:\n", a)

print("\nShape:", a.shape)

print("Dtype:", a.dtype)

print("Total Sum:", np.sum(a))

print("\nRow-wise Means:")
print(np.mean(a, axis=1))

print("\nColumn-wise Standard Deviations:")
print(np.std(a, axis=0))

m = np.max(a)

p = np.unravel_index(np.argmax(a), a.shape)

print("\nGlobal Max:", m)

print("Position:", p)

c = np.sum(a > 50)

print("\nCount of elements > 50:", c)
```
