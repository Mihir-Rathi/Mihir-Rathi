```python
import numpy as np
import time

x = np.random.randint(1, 100, 1000000)

start = time.time()

result = []
for i in range(len(x)):
    result.append(x[i]**2 + 2*x[i] + 1)

end = time.time()

print("Loop Time:", end - start)

start = time.time()

result2 = x**2 + 2*x + 1

end = time.time()

print("Vectorized Time:", end - start)

print("\nSame Result?")
print(np.allclose(result, result2))
```

### t1.py

```python
import numpy as np

def stats_report(arr):
    arr = np.array(arr)

    shape = arr.shape
    minimum = np.min(arr)
    maximum = np.max(arr)
    mean = np.mean(arr)
    median = np.median(arr)
    std = np.std(arr)
    p25 = np.percentile(arr, 25)
    p75 = np.percentile(arr, 75)

    max_pos = np.unravel_index(np.argmax(arr), arr.shape)

    print("Shape :", shape)
    print("Minimum :", minimum)
    print("Maximum :", maximum)
    print("Mean :", mean)
    print("Median :", median)
    print("Standard Deviation :", std)
    print("25th Percentile :", p25)
    print("75th Percentile :", p75)
    print("Global Maximum Position (row, col) :", max_pos)

arr = np.array([
    [10, 25, 30],
    [45, 12, 18],
    [7, 60, 22]
])

stats_report(arr)
```

```python
import numpy as np

marks = np.random.randint(35, 101, (30, 5))

avg_marks = np.mean(marks, axis=1)

ranks = avg_marks.argsort()[::-1].argsort() + 1

top_5 = np.argsort(avg_marks)[-5:][::-1]

bottom_5 = np.argsort(avg_marks)[:5]

correlation = np.corrcoef(marks[:, 0], marks[:, 1])

print("Marks:\n", marks)

print("\nAverage Marks Per Student:\n", avg_marks)

print("\nRanks:\n", ranks)

print("\nTop 5 Students Index:\n", top_5)

print("\nBottom 5 Students Index:\n", bottom_5)

print("\nCorrelation Between Subject 1 and Subject 2:\n", correlation)
```

### t5.py

```python
import numpy as np

returns = np.random.uniform(-0.1, 0.2, (12, 5))

mean_return = np.mean(returns, axis=0)

cov_matrix = np.cov(returns, rowvar=False)

weights = np.array([0.2] * 5)

portfolio_variance = weights @ cov_matrix @ weights

asset_risk = np.std(returns, axis=0)

highest_risk = np.argmax(asset_risk)

lowest_risk = np.argmin(asset_risk)

print("Monthly Returns:\n", returns)

print("\nMean Return Per Asset:\n", mean_return)

print("\nCovariance Matrix:\n", cov_matrix)

print("\nPortfolio Variance:\n", portfolio_variance)

print("\nHighest Risk Asset Index:", highest_risk)

print("Lowest Risk Asset Index:", lowest_risk)
```
