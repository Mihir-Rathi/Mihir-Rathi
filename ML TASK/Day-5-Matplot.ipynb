### t1.py

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y = x**2

plt.plot(x, y, linestyle='--', color='green')
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.title("Squares of Numbers from 1 to 10")

plt.show()
```

### t2.py

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)

plt.plot(x, x**2, label='y = x^2')
plt.plot(x, x**3, label='y = x^3')

plt.legend()

plt.show()
```

### t3.py

```python
import numpy as np
import matplotlib.pyplot as plt

height = np.array([150, 155, 160, 165, 170, 175, 180, 185])
weight = np.array([50, 55, 60, 65, 70, 75, 80, 85])

plt.scatter(height, weight, color='purple', marker='^')

plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight")

plt.show()
```

### t4.py

```python
import matplotlib.pyplot as plt

countries = ["India", "USA", "China", "Brazil", "Canada"]
population = [1420, 340, 1410, 216, 40]

colors = ["orange", "blue", "red", "green", "purple"]

plt.barh(countries, population, color=colors)

plt.xlabel("Population (millions)")
plt.title("Population of Countries")

plt.show()
```

### t5.py

```python
import numpy as np
import matplotlib.pyplot as plt

ages = np.random.randint(0, 81, 500)

plt.hist(ages, bins=10, color='skyblue', edgecolor='black')

plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Ages")

plt.show()
```

### t6.py

```python
import matplotlib.pyplot as plt

activities = ["Sleep", "Work", "Study", "Leisure"]
hours = [8, 8, 5, 3]

explode = [0, 0, 0.2, 0]

colors = ["skyblue", "orange", "green", "purple"]

plt.pie(hours,labels=activities,colors=colors,explode=explode,autopct='%1.1f%%')

plt.title("Daily Time Allocation")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y = x**2

plt.plot(x, y, linestyle='--', color='green')

plt.title("Squares of Numbers")
plt.xlabel("Numbers")
plt.ylabel("Squares")

plt.grid(True)

plt.xlim(0, 12)

plt.show()
```

### t8.py

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 6)

fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0, 0].plot(x, x**2)
ax[0, 0].set_title("Line Plot")

ax[0, 1].scatter(x, x**2, color='red')
ax[0, 1].set_title("Scatter Plot")

ax[1, 0].bar(x, x**2, color='green')
ax[1, 0].set_title("Bar Chart")

data = np.random.randint(1, 100, 100)

ax[1, 1].hist(data, bins=10, color='skyblue', edgecolor='black')
ax[1, 1].set_title("Histogram")

plt.tight_layout()

plt.savefig("subplot.jpg",dpi=300)
plt.show()
```
