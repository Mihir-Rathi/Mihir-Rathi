### t1.py

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

plt.figure(figsize=(8, 6))
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", s=80)
plt.title("Penguin Bill Dimensions by Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.legend(title="Species")
plt.tight_layout()
plt.savefig("penguins_scatter.png", dpi=150)
plt.show()
```

### t2.py

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

plt.figure(figsize=(10, 6))
sns.lineplot(data=flights, x="year", y="passengers")
plt.title("Number of Passengers Over Years")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.tight_layout()
plt.savefig("flights_line.png", dpi=150)
plt.show()
```

### t3.py

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

plt.figure(figsize=(10, 6))
sns.histplot(data=titanic, x="age", hue="survived", kde=True, bins=30)
plt.title("Titanic Passenger Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.tight_layout()
plt.savefig("titanic_age_hist.png", dpi=150)
plt.show()
```

### t4.py

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

corr = tips.corr(numeric_only=True)

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="viridis", fmt=".2f", linewidths=0.5, square=True)
plt.title("Correlation Matrix - Tips Dataset")
plt.tight_layout()
plt.savefig("tips_correlation_heatmap.png", dpi=150)
plt.show()
```

### t6.py

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title("Fare Distribution by Passenger Class on Titanic")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.tight_layout()
plt.savefig("titanic_fare_boxplot.png", dpi=150)
plt.show()
```

### t7.py

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x="pclass")
plt.title("Number of Passengers by Class on Titanic")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("titanic_pclass_count.png", dpi=150)
plt.show()
```

### t8.py

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.pairplot(iris, hue="species", diag_kind="kde", height=2.5)
plt.suptitle("Iris Dataset Pairplot by Species", y=1.02)
plt.tight_layout()
plt.savefig("iris_pairplot.png", dpi=150)
plt.show()

print("Analysis: Setosa is the most easily separable species.")
print("It forms distinct clusters across all feature pairs, especially in petal measurements.")
print("Versicolor and Virginica show some overlap, making them harder to distinguish.")
```
